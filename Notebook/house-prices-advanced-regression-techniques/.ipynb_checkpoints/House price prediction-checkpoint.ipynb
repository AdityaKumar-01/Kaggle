```python
import pandas as pd
df_train = pd.read_csv("../../DATA/house-prices-advanced-regression-techniques/", index_col="Id")
```

```python
df_train.head()
```

```python
df_test = pd.read_csv("./test.csv", index_col = 'Id')
```

```python
df_test.head()
```

```python
target = df_train['SalePrice']
df_train.drop('SalePrice',axis=1, inplace=True)
```

```python
df_train['training_set']=True
df_test['training_set']= False
```

```python
df_full = pd.concat([df_train, df_test])
```

```python
df_full = df_full.interpolate()
df_full= pd.get_dummies(df_full)
```

```python
df_train = df_full[df_full['training_set']==True]
df_train.drop('training_set', axis=1, inplace=True)

df_test = df_full[df_full['training_set']==False]
df_test.drop('training_set', axis=1, inplace=True)
```

```python
from sklearn.ensemble import RandomForestRegressor
rf  = RandomForestRegressor(n_estimators=100, n_jobs = -1)
```

```python
rf.fit(df_train, target)
```

```python
preds = rf.predict(df_test)
```

```python
my_sub= pd.DataFrame({'Id':df_test.index, 'SalePrice' : preds})
```

```python
my_sub.to_csv('./submission.csv', index=False)
```
